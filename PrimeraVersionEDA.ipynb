```python
import numpy as np
def impute_mixed(
    df: pd.DataFrame,
    numeric_strategy: str = "mean",    
    categorical_strategy: str = "mode",
    fill_value: str = "SIN_DATO",
    report: bool = True,
    save_csv: str = None
    ) -> pd.DataFrame:

    df_imp = df.copy()

    num_cols = df_imp.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = df_imp.select_dtypes(include=["object", "category"]).columns.tolist()

    if report:
        print("Nulos ANTES de imputar:")
        print(" - Numéricas:", int(df_imp[num_cols].isna().sum().sum()) if num_cols else 0)
        print(" - Categóricas:", int(df_imp[cat_cols].isna().sum().sum()) if cat_cols else 0)

    if num_cols:
        if numeric_strategy == "median":
            df_imp[num_cols] = df_imp[num_cols].fillna(df_imp[num_cols].median())
        else:  
            df_imp[num_cols] = df_imp[num_cols].fillna(df_imp[num_cols].mean())

    if cat_cols:
        if categorical_strategy == "mode":
            for c in cat_cols:
                if df_imp[c].isna().any():
                    moda = df_imp[c].mode(dropna=True)
                    fill = moda.iloc[0] if not moda.empty else fill_value
                    df_imp[c] = df_imp[c].fillna(fill)
        else:  
            df_imp[cat_cols] = df_imp[cat_cols].fillna(fill_value)

    if report:
        print("\nNulos DESPUÉS de imputar:")
        print(" - Numéricas:", int(df_imp[num_cols].isna().sum().sum()) if num_cols else 0)
        print(" - Categóricas:", int(df_imp[cat_cols].isna().sum().sum()) if cat_cols else 0)
    if save_csv:
        df_imp.to_csv("Bases_Datos/dataset_imputado.csv", index=False, encoding="utf-8-sig")

    return df_imp


df_imputado = impute_mixed(df,
                           numeric_strategy="mean",     
                           categorical_strategy="mode",  
                           fill_value="SIN_DATO",
                           report=True,
                           save_csv="dataset_imputado.csv" )
```

```python
def timeskip(datetime: pd.DataFrame, delta : pd.Timedelta = pd.Timedelta('0 days 01:00:00')):
    timeskips = []
    for i in range(1, len(datetime)):
        i_delta = datetime[i] - datetime[i-1]
        
        if i_delta != delta:
            timeskips.append(f"[{i-1} : {i}] - Delta: {i_delta}")


estaciones = df.groupby("Estacion")

timeskip_df = pd.DataFrame()
for name, group in estaciones:
    timeskip_df[name] = timeskip(sorted(group["Fecha"]))

timeskip_df
```
